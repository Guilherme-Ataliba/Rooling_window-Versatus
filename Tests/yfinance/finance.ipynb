```python
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
sugar = yf.Ticker("SB=F")
coffee = yf.Ticker("KC=F")
```

```python
sugar_history = sugar.history(period="1y")
sugar_history = sugar_history.asfreq("B")
sugar_history.head()
```

```python
coffee_history = coffee.history(period="1y")
coffee_history = coffee_history.asfreq("B")
coffee_history.head()
```

```python
def plot_open_and_volume(data, title=""):
    fig, axs = plt.subplots(1, 2, figsize=(16, 4))

    axs[0].plot(data.index, data["Open"])
    axs[0].set_title("Open")


    axs[1].plot(data.index, data["Volume"])
    axs[1].set_title("Volume")

    plt.suptitle(title)

    plt.show()
```

```python
plot_open_and_volume(sugar_history, "Sugar (SB=F)")
plot_open_and_volume(coffee_history, "Coffee KC=F")
```

# Decomposition

```python
from statsmodels.tsa.seasonal import seasonal_decompose

def plot_decomposicao(data, title):
    decomposicao = seasonal_decompose(data)
    fig, axs = plt.subplots(4, 1, sharex=True, figsize=(14, 8))
    decomposicao.observed.plot(ax=axs[0], title="Dados Observados")
    decomposicao.trend.plot(ax = axs[1], title="Tendncia")
    decomposicao.seasonal.plot(ax = axs[2], title="Sazonalidade")
    decomposicao.resid.plot(ax=axs[3], title="Resduos")
    plt.suptitle(title)
```

```python
plot_decomposicao(sugar_history["Open"], "Sugar Decomposition")
```

```python
plot_decomposicao(coffee_history["Open"], "Coffee Decomposition")
```

# Statistical Analysys

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf


fig, axs = plt.subplots(1, 2, figsize=(20, 5))
plot_acf(sugar_history["Open"], lags=30, ax=axs[0])
plot_pacf(sugar_history["Open"], lags=30, ax=axs[1])
plt.suptitle("Sugar")
plt.show()


fig, axs = plt.subplots(1, 2, figsize=(20, 5))
plot_acf(coffee_history["Open"], lags=30, ax=axs[0])
plot_pacf(coffee_history["Open"], lags=30, ax=axs[1])
plt.suptitle("Coffee")
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller, kpss

def stationary_tests(data):
    adf = adfuller(data)
    kp = kpss(data)
    if adf[1] > 0.05:
        print(f"ADF: {adf[1]} No estacionrio")
    else:
        print(f"ADF: {adf[1]} Estacionrio")

    if kp[1] < 0.05:
        print(f"kpss: {kp[1]} No estacionrio")
    else:
        print(f"kpss: {kp[1]} Estacionrio")
```

```python
print("-="*20, "Sugar", "-="*20)
stationary_tests(sugar_history["Open"])
print("-="*20, "Coffee", "-="*20)
stationary_tests(coffee_history["Open"])
print("-="*44)
```

# ARIMA

```python
sugar_decomposition = seasonal_decompose(sugar_history["Open"])
sugar_decomposition.seasonal[0:20].plot()
```

```python
divisao = int(len(sugar_history)*0.8)

X_train = sugar_history.index[:divisao]
X_test = sugar_history.index[divisao:]

treino = sugar_history["Open"][:divisao].asfreq("B")
teste = sugar_history["Open"][divisao:].asfreq("B")
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

arima = SARIMAX(treino, order=(2, 1, 2), seasonal_order=(2,1,1,3)).fit()
```

```python
train_pred = arima.predict(2, len(treino)-1)
plt.plot(sugar_history.index[:divisao], treino, label="Train Actual")
plt.plot(X_train[2:], train_pred, label="Train predicted")
plt.legend()
```

```python
test_pred = arima.predict(len(treino), len(treino)+len(teste)-1)
plt.plot(sugar_history.index[divisao:], teste, label="Test Actual")
plt.plot(X_test, test_pred, label="Test Predicted")
plt.legend()
```
